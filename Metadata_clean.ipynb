```python
import pandas as pd
import numpy as np
ad = pd.read_csv("movies_metadata.csv")
ad
```

```python
ad.drop(['adult','homepage','video','revenue','belongs_to_collection','poster_path','budget','id','overview','release_date','status','tagline'], axis=1, inplace=True)
```

```python
ad=ad.drop(['production_countries','production_companies','spoken_languages','popularity'], axis=1)
```

```python
ad[ad['runtime'].isnull()]['runtime'].value_counts(dropna=False)
```

```python
mean_runtime = ad['runtime'].value_counts().mean()
ad['runtime'].fillna(mean_runtime, inplace=True)
```

```python
ad=ad.sort_values(by=['vote_average'], ascending=False)
```

```python
ad['vote_average'].dropna(inplace=True)
```

```python
ad=ad.reset_index()
```

```python
ad=ad[ad['vote_average'].notnull()]
```

```python
ad=ad[ad['original_language'] == 'en']
ad=ad.reset_index()
ad
```

```python
ad=ad[~ad['genres'].str.contains('\[]')]
ad.drop(['level_0','index'],axis=1, inplace=True)
```

```python
ad=ad.reset_index()
```

```python
ad
```

```python
ad['genres']=ad['genres'].apply(lambda genres: eval(genres))



```

```python
ad['genres']=ad['genres'].apply(lambda genres: genres[0]['name'])
```

```python
ad.drop('index', axis=1, inplace=True)
```
