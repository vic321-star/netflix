```python
import pandas as pd
import numpy as np
```

```python
rd = pd.read_csv("movies_metadata.csv")
low_memory=False
rd.nunique()
```

```python
rd=rd[rd['original_language'] == 'en']
```

```python
rd=rd[rd['runtime'] >= 90]
rd=rd[rd['vote_count'] >= 500]
```

```python
rd.reset_index(drop=True,inplace=True)
rd
```

```python
ad=rd[rd['genres'].apply(lambda genres: "Action" in genres)]
bd=rd[rd['genres'].apply(lambda genres: "Adventure" in genres)]
```

```python
rd = pd.merge(ad,bd)
rd
```
