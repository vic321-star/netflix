```python
rd = pd.read_csv("movies_metadata.csv")
low_memory=False
```

```python
rd=rd[rd['original_language'] == 'en']
```

```python
rd=rd[rd['runtime'] >= 140]
rd=rd[rd['vote_count'] >= 1000]
```
